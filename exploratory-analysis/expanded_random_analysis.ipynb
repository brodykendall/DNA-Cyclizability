```python
import mpl_scatter_density # adds projection='scatter_density'
from matplotlib.colors import LinearSegmentedColormap
```

```python
# plt.figure(figsize=(6,4))
# fig = plt.figure(figsize=(4,3))
# fig = plt.figure(figsize=(5,3))
fig = plt.figure(figsize=(4,3))
ax = fig.add_subplot(1, 1, 1, projection="scatter_density")

from scipy.stats import gaussian_kde

white_viridis = LinearSegmentedColormap.from_list('white_viridis', [
    (0, '#ffffff'),
    (1e-20, '#440053'),
    (0.2, '#404388'),
    (0.4, '#2a788e'),
    (0.6, '#21a784'),
    (0.8, '#78d151'),
    (1, '#fde624'),
], N=256)

# Calculate the point density
x=combined_smoothC0_df.loc[~combined_smoothC0_df["order1_smoothC0"].isna()]["order1_smoothC0"]
y=combined_smoothC0_df.loc[~combined_smoothC0_df["order1_smoothC0"].isna()]["order2_smoothC0"]
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)

# Sort the points by density, so that the densest points are plotted last
idx = z.argsort()
x, y, z = x.iloc[idx], y.iloc[idx], z[idx]

# fig, ax = plt.subplots()
# dense_scatter=ax.scatter(x, y, c=z, s=50)
dense_scatter=ax.scatter_density(x, y, cmap=white_viridis)
fig.colorbar(dense_scatter)
# plt.show()


# plt.scatter(x, y, c=z)

# plt.scatter(combined_smoothC0_df["order1_smoothC0"], combined_smoothC0_df["order2_smoothC0"], alpha=0.1)
# plt.xlabel("Order 1 Smooth C0")
# plt.ylabel("Order 2 Smooth C0")

plt.show()
```
