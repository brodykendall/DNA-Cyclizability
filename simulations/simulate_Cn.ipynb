```python
def find_c0(dat, aa, C26_col, C29_col, C31_col):
  mat = np.empty((3,3), float)
  k = 2*np.pi/10.4
  n = array([26, 29, 31])
  mat[0:3, 0] = 1
  mat[0:3, 1] = np.sin(n*k)
  mat[0:3, 2] = np.cos(n*k)
  mat[0, 1:3] = mat[0, 1:3]*aa[0]
  mat[1, 1:3] = mat[1, 1:3]*aa[1]
  mat[2, 1:3] = mat[2, 1:3]*aa[2]
  inv_mat = np.linalg.inv(mat)
  c0A1A2 = array(np.matmul(dat[[C26_col, C29_col, C31_col]], np.transpose(inv_mat))).astype(float)
  c0Aphi = c0A1A2
  c0Aphi[:,0] = c0A1A2[:,0]
  c0Aphi[:,1] = np.sqrt(c0A1A2[:,1]**2 + c0A1A2[:,2]**2)
  c0Aphi[:,2] <- np.sign(c0A1A2[:,2]) * np.arccos(c0A1A2[:,1]/c0Aphi[:,1])
  return c0Aphi[:,0]
```

```python
A29_ratios = A31_ratios = [x/10 for x in range(5,16)]
sim1_corrs = np.zeros((len(A29_ratios),len(A31_ratios)))
sim2_corrs = np.zeros((len(A29_ratios),len(A31_ratios)))
sim3_corrs = np.zeros((len(A29_ratios),len(A31_ratios)))
for A29_ratio_idx, A29_ratio in enumerate(A29_ratios):
    for A31_ratio_idx, A31_ratio in enumerate(A31_ratios):
        cur_aa = [1, 1/A29_ratio, 1/A31_ratio]
        
        cur_C0_sim1 = find_c0(pd.DataFrame({"C26_sim1": C26_sim1,
                                            "C29_sim1": C29_sim1,
                                            "C31_sim1": C31_sim1}), 
                                            cur_aa, "C26_sim1", "C29_sim1", "C31_sim1")
        cur_C0_sim2 = find_c0(pd.DataFrame({"C26_sim2": C26_sim2,
                                            "C29_sim2": C29_sim2,
                                            "C31_sim2": C31_sim2}), 
                                            cur_aa, "C26_sim2", "C29_sim2", "C31_sim2")
        cur_C0_sim3 = find_c0(pd.DataFrame({"C26_sim3": C26_sim3,
                                            "C29_sim3": C29_sim3,
                                            "C31_sim3": C31_sim3}), 
                                            cur_aa, "C26_sim3", "C29_sim3", "C31_sim3")
        
        sim1_corrs[A29_ratio_idx, A31_ratio_idx] = np.corrcoef(cur_C0_sim1, C0_sim)[0,1]
        sim2_corrs[A29_ratio_idx, A31_ratio_idx] = np.corrcoef(cur_C0_sim2, C0_sim)[0,1]
        sim3_corrs[A29_ratio_idx, A31_ratio_idx] = np.corrcoef(cur_C0_sim3, C0_sim)[0,1]
```
