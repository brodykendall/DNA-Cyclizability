```python
import pandas as pd
import numpy as np
```

```python
nuc_library = pd.read_csv("/Users/Brody1/Dropbox/Northwestern/DNA_Cyclizability/cycle1.txt",delimiter = ",")
```

```python
np.random.seed(1)
# Random vector with length equal to the number of sequences, each value in [0, 49]:
nuc_library_rotation_values = np.random.randint(0, 50, size=nuc_library.shape[0])
nuc_library_shifted_sequences = ["".join([nuc_library["Sequence"][i][nuc_library_rotation_values[i]:], nuc_library["Sequence"][i][:nuc_library_rotation_values[i]]]) for i in range(nuc_library.shape[0])]

nuc_library["Sequence"] = nuc_library_shifted_sequences
nuc_library["rotation_values"] = nuc_library_rotation_values

nuc_library.to_csv("/Users/Brody1/Dropbox/Northwestern/DNA_Cyclizability/data/Created/nuc_rotated.csv", index=False)
```
