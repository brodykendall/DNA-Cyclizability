```python
import pandas as pd
```

```python
sac2_positions_file = "/Users/Brody1/Dropbox/Northwestern/DNA_Cyclizability/sgdSAC2_ncp.txt"
sac2_positions = pd.read_csv(sac2_positions_file, delim_whitespace=True, header=None, names=['chrID', 'position', 'ncp_score', 'ncp_ratio'])
```

```python
sac2_positions
```

```python
sac2_positions_bed_file = "/Users/Brody1/Dropbox/Northwestern/DNA_Cyclizability/sgdSAC2_ncp_reformat.bed"

with open(sac2_positions_bed_file, "w") as f_out:
    for idx, line in sac2_positions.iterrows():
        chromosome = line["chrID"]
        position = line["position"]
        f_out.write(f"{chromosome}\t{position}\t{position}\n")
```

## Converted to SacCer3 using liftover here, saved as sgdSAC3_ncp.bed

```python
sac3_positions_bed_file = "/Users/Brody1/Dropbox/Northwestern/DNA_Cyclizability/sgdSAC3_ncp.bed"
sac3_positions = pd.read_csv(sac3_positions_bed_file, sep="\t", header=None)
sac3_positions = sac3_positions.loc[:,[0,1]]
sac3_positions = sac3_positions.rename(columns={0: "chrID", 1: "position"})
```

```python
sac3_positions
```

```python
# Deleted in new:
#         Sequence intersects no chains
#     Partially deleted in new:
#         Sequence insufficiently intersects one chain
#     Split in new:
#         Sequence insufficiently intersects multiple chains
#     Duplicated in new:
#         Sequence sufficiently intersects multiple chains
#     Boundary problem:
#         Missing start or end base in an exon

# Deleted in new:
# chrXI	70486	70486

sac2_positions_dropped = sac2_positions.loc[~((sac2_positions['chrID'] == 'chrXI') & (sac2_positions['position'] == 70486))].reset_index()
```

```python
sac2_positions_dropped
```

```python
sac3_positions["ncp_score"] = sac2_positions_dropped["ncp_score"]
sac3_positions["ncp_ratio"] = sac2_positions_dropped["ncp_ratio"]
```

```python
sac3_positions
```

```python
sac3_positions_file = "/Users/Brody1/Dropbox/Northwestern/DNA_Cyclizability/sgdSAC3_ncp.txt"
sac3_positions.to_csv(sac3_positions_file, index=False, sep="\t", header=None)
```
