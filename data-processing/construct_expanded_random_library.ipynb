```python
import pandas as pd
import numpy as np
```

```python
random_library = pd.read_csv("/Users/Brody1/Dropbox/Northwestern/DNA_Cyclizability/cycle3.txt",delimiter = ",")
```

```python
random_library.shape
```

```python
order1 = list(range(random_library.shape[0]))
order2 = list(range(random_library.shape[0]))
np.random.seed(50)
np.random.shuffle(order1)
np.random.shuffle(order2)
```

```python
print(order1[:5], order2[:5])
```

```python
long_seq1 = ""
long_seq2 = ""
for i in range(random_library.shape[0]):
    seq1 = random_library.loc[order1[i], "Sequence"]
    seq2 = random_library.loc[order2[i], "Sequence"]
    long_seq1 += seq1
    long_seq2 += seq2
# for seq in random_library["Sequence"]:
#     long_seq += seq
```

```python
long_seq1
```

```python
all_seq_list1 = []
all_seq_list2 = []
for i in range(50, len(long_seq1)+1):
    all_seq_list1.append(long_seq1[(i-50):i])
    all_seq_list2.append(long_seq2[(i-50):i])
```

```python
expanded_random1 = pd.DataFrame({"sequence":all_seq_list1})
expanded_random2 = pd.DataFrame({"sequence":all_seq_list2})
```

```python
expanded_random1
```

```python
j = 0
k = 0
expanded_random1["C26"] = np.nan
expanded_random1["C29"] = np.nan
expanded_random1["C31"] = np.nan
expanded_random1["C0"] = np.nan
expanded_random1["orig_idx"] = np.nan

expanded_random2["C26"] = np.nan
expanded_random2["C29"] = np.nan
expanded_random2["C31"] = np.nan
expanded_random2["C0"] = np.nan
expanded_random2["orig_idx"] = np.nan
for i in range(expanded_random1.shape[0]):
    if expanded_random1.loc[i, "sequence"] == random_library.loc[order1[j], "Sequence"]:
        expanded_random1.loc[i, "C26"] = random_library.loc[order1[j], "n=26"]
        expanded_random1.loc[i, "C29"] = random_library.loc[order1[j], "n=29"]
        expanded_random1.loc[i, "C31"] = random_library.loc[order1[j], "n=31"]
        expanded_random1.loc[i, "C0"] = random_library.loc[order1[j], "C0"]
        expanded_random1.loc[i, "orig_idx"] = int(order1[j])
        j+=1
    if expanded_random2.loc[i, "sequence"] == random_library.loc[order2[k], "Sequence"]:
        expanded_random2.loc[i, "C26"] = random_library.loc[order2[k], "n=26"]
        expanded_random2.loc[i, "C29"] = random_library.loc[order2[k], "n=29"]
        expanded_random2.loc[i, "C31"] = random_library.loc[order2[k], "n=31"]
        expanded_random2.loc[i, "C0"] = random_library.loc[order2[k], "C0"]
        expanded_random2.loc[i, "orig_idx"] = int(order2[k])
        k+=1
    
```

```python
print(sum(~np.isnan(expanded_random1["C26"])), sum(~np.isnan(expanded_random2["C26"])))
```
