```python
import keras
import pandas as pd
import numpy as np
from numpy import array
```

```python
def dnaOneHot(sequence):
    seq_array = array(list(sequence))
    code = {"A": [0], "C": [1], "G": [2], "T": [3], "N": [4],
            "a": [0], "c": [1], "g": [2], "t": [3], "n": [4]}
    onehot_encoded_seq = []
    for char in seq_array:
        onehot_encoded = np.zeros(5)
        onehot_encoded[code[char]] = 1
        onehot_encoded_seq.append(onehot_encoded[0:4])
    return onehot_encoded_seq
```

```python
X = []
for sequence_nt in df[sequence_column_name]:
    X.append(dnaOneHot(sequence_nt))
X = array(X)
X = X.reshape((X.shape[0],50,4,1))
X_reverse = np.flip(X,[1,2])
```

```python
predictions = np.zeros([X.shape[0], 10])
for i in range(1,11):
    cur_model = keras.models.load_model(f"/Users/Brody1/Dropbox/Northwestern/DNA_Cyclizability/benchmarks/ir_lstm_{model_identifier}_{training_data_name}_{i}.h5")
    # cur_model_pred_norm = cur_model.predict(X)
    cur_model_pred = cur_model.predict(X)
    cur_model_pred_reverse = cur_model.predict(X_reverse)
    cur_model_pred = (cur_model_pred+cur_model_pred_reverse)/2
    # cur_model_pred = cur_model_pred_norm*target_std + target_mean
    predictions[:,i-1] = cur_model_pred.flatten()
```

```python
avg_predictions = np.mean(predictions, axis=1)
```

```python
for target_column_name in target_column_name_list:
    corr_df = pd.DataFrame({"target": df[target_column_name], "predicted": avg_predictions})
    print(f"{training_column_name} prediction correlation on {data_name}, {target_column_name}: {corr_df.corr().iloc[0,1]}")
    # print(f"{training_column_name} prediction correlation on {data_name}, {target_column_name}: {np.corrcoef(avg_predictions, df[target_column_name])[0,1]}")
```
